### Executive Summary


The Supply Chain Data Analysis project aimed to develop a machine learning model to predict whether a new global health commodity shipment would deliver on its scheduled date or not. This information would be valuable to supply chain analysts who could use it to plan for potential delays and take proactive steps to mitigate them.

The project analyzed over 10,000 shipment records from various sources, including USAID's Project Procurement Request (PQR) system. The data was analyzed to identify patterns and correlations between shipment attributes and delivery delays. A Logistic Regression and Random Forest machine learning models are developed to predict delivery delays. The model achieved an accuracy of 87% and 99% respectively on the test data.

The development of this machine learning model has the potential to significantly improve the efficiency of global health supply chains. By being able to predict delivery delays, supply chain analysts can take proactive steps to mitigate them, such as expediting shipments or finding alternative routes.

The model could be further developed to provide more granular predictions, such as predicting the likelihood of delays at different stages of the shipment process. Additionally, the model could be used to optimize shipment planning across stages, such as understanding manufacturing delays, Shipment delays, recommending ideal carriers by route among other opportunities.


### Introduction

The Supply Chain Shipment Pricing Dataset is a valuable resource for researchers and practitioners interested in understanding and optimizing supply chain shipment pricing. It contains over 10,000 shipment records from various sources, including USAID's Project Procurement Request (PQR) system. The dataset provides comprehensive information on shipment characteristics, including origin and destination countries, commodity types, transportation modes, and associated costs.




### Project Objectives


The primary objective of this project was to analyze the data to provide insights and develop a machine learning model to predict whether a new global health commodity shipment would deliver on its scheduled date or not. This information would be valuable to supply chain analysts who could use it to plan for potential delays and take proactive steps to mitigate them.

### Data Analysis

The data was analyzed to identify patterns and correlations between shipment attributes and delivery delays. Some of the key findings included:

Larger order sizes were more likely to be delayed.
Shipments of certain product types, such as pharmaceuticals, were more likely to be delayed.
Shipments that were routed through certain regions, such as Africa, were more likely to be delayed.

### Percentage of On-Time and Delayed Deliveries by Top 20 Vendors

```python
top_vendors = df['Vendor'].value_counts().nlargest(20).index
df_top_vendors = df[df['Vendor'].isin(top_vendors)]

# Create a contingency table
contingency_table = pd.crosstab(df_top_vendors['Vendor'], df_top_vendors['on_time_delivery'], margins=True, margins_name="Total")

# Calculate percentage
contingency_table_percentage = contingency_table.div(contingency_table['Total'], axis=0) * 100
ax = contingency_table_percentage.drop('Total', axis=1).plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Percentage of On-Time and Delayed Deliveries by Top 20 Vendors')
plt.xlabel('Vendor')
plt.ylabel('Percentage')
plt.legend(title='On-Time Delivery', bbox_to_anchor=(1, 1))

plt.show()
```

### Question 6: Examine if there are any trends or patterns in on-time delivery over time.  This could involve looking at monthly or quarterly variations to identify any seasonal trends or improvements

```python
'''On-Time Delivery Over Time:

Examine if there are any trends or patterns in on-time delivery over time. 
This could involve looking at monthly or quarterly variations to identify any seasonal trends or improvements.'''

df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'])  # Convert to datetime for time-based analysis
df['Year-Month'] = df['Delivered to Client Date'].dt.to_period('M')  # Extract Year-Month for monthly analysis

# Create a contingency table
contingency_table_time = pd.crosstab(df['Year-Month'], df['on_time_delivery'], margins=True, margins_name="Total")

# Calculate percentage
contingency_table_time_percentage = contingency_table_time.div(contingency_table_time['Total'], axis=0) * 100

# Plot the line plot
ax = contingency_table_time_percentage.drop('Total', axis=1).plot(kind='line', marker='o', figsize=(12, 6))
plt.title('On-Time Delivery Percentage Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Percentage')
plt.legend(title='On-Time Delivery', bbox_to_anchor=(1, 1))

plt.show()
```

#### Answer: Looks like most shipment were deliverd on time and there was sharp drop in 2010 and was close to 50% for 2 years.The trend after that has inclined. We can investigate further to analyze sharp dropis in year 2010-2011.

```python
'''On-Time Delivery by Fulfillment Mode:

Investigate whether the fulfillment mode (e.g., Direct Drop, CIP, EXW) affects on-time delivery.
Different fulfillment modes may have varying levels of efficiency and reliability.'''



# Create a contingency table
contingency_table_fulfillment = pd.crosstab(df['Fulfill Via'], df['on_time_delivery'], margins=True, margins_name="Total")

# Calculate percentage
contingency_table_fulfillment_percentage = contingency_table_fulfillment.div(contingency_table_fulfillment['Total'], axis=0) * 100

# Plot the bar plot
ax = contingency_table_fulfillment_percentage.drop('Total', axis=1).plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('On-Time Delivery Percentage by Fulfillment Mode')
plt.xlabel('Fulfillment Mode')
plt.ylabel('Percentage')
plt.legend(title='On-Time Delivery', bbox_to_anchor=(1, 1))

plt.show()

```
